```python
# d
import numpy as np
# N > d+1
X = [[2,4,6],[1,3,5],[1,4,2],[3,5,2]]
X = np.asarray(X)
X_cross = np.linalg.pinv(X)
XX_cross = np.dot(X,X_cross)
trace = np.trace(XX_cross)
rank = np.linalg.matrix_rank(X)

print("trace: ",trace)
print("rank: ",rank)
```
